```python
import pandas as pd

df = pd.read_csv("ecommerce_reviews.csv")
df.head(5)
```

```python
print("Total Rows In The Dataset:",df.shape[0])
print("Total Columns In The Dataset:",df.shape[1],"\n")

print("Columns Names: ",df.columns,"\n")

print(f"Date Range: {df['Review_Date'].min()} to {df['Review_Date'].max()}")
```

```python
print("Missing Values In The Dataset:")
print(df.isnull().sum())
```

```python
print("Data Info:")
print(df.info())
```

```python
print("Unique Values:")
print(df.nunique())
```
